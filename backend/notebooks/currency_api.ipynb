```python
import pandas as pd
import datetime
from datetime import timedelta, date
import time
import requests
from tqdm import tqdm  # For progress tracking

# Generate monthly dates from 2022 to 2025
dates = pd.date_range(start='2022-01-01', end='2025-03-01', freq='MS').strftime('%Y-%m-%d').tolist()
print(f"Will fetch data for {len(dates)} dates")

# Define currencies we're interested in
target_currencies = ['BND', 'MMK', 'KHR', 'USD', 'IDR', 'LAK', 'MYR', 'PHP', 'SGD', 'THB', 'VND']

# Create a list to store DataFrames
df_list = []

# Use tqdm for progress tracking
for date_str in tqdm(dates, desc="Fetching currency data"):
    try:
        # Using USD as base currency instead of CHF for easier comparison
        url = f'https://www.xe.com/currencytables/?from=USD&date={date_str}'
        print(f"Fetching data from: {url}")
        
        # Fetch the table
        tables = pd.read_html(url)
        if tables and len(tables) > 0:
            dfs = tables[0]
            dfs['Date'] = date_str
            
            # Append to our list
            df_list.append(dfs)
            
            # Be nice to the server
            time.sleep(5)  # Add a delay to avoid getting blocked
            
    except Exception as e:
        print(f"Error fetching data for {date_str}: {e}")
        time.sleep(10)  # Wait longer after an error

# Combine all dataframes
if df_list:
    df = pd.concat(df_list, ignore_index=True)
    
    # Now filter for specific currencies
    # The column name in the XE table might be 'Currency Code' not 'Currency'
    # Let's check the actual column names
    print("DataFrame columns:", df.columns.tolist())
    
    # Filter based on the correct column name
    # Assuming it's 'Currency Code' - adjust if it's different
    if 'Currency Code' in df.columns:
        df_filtered = df[df['Currency Code'].isin(target_currencies)]
    elif 'Currency' in df.columns:
        df_filtered = df[df['Currency'].isin(target_currencies)]
    else:
        print("Could not find currency column. Printing first few rows to debug:")
        print(df.head())
        df_filtered = df  # Just use all data if we can't filter
    
    # Save the filtered data
    df_filtered.to_csv('/Users/lishuyao/Documents/NUS/MODS/Y3S2/BT4301/tourism-consultancy/extracted_data/southeast_asia_currency_rates.csv', index=False)
    print(f"Data saved to southeast_asia_currency_rates.csv")
    
    # Also display a summary
    print(f"Saved data for {df_filtered['Date'].nunique()} dates and {len(target_currencies)} currencies")
else:
    print("No data was collected")
```

```python
import pandas as pd
import datetime
from datetime import timedelta
import time
import requests
from tqdm import tqdm

# Function to generate all dates between two dates
def daterange(start_date, end_date):
    for n in range(int((end_date - start_date).days) + 1):
        yield start_date + timedelta(n)

# Define the date range for daily data extraction
start_date = datetime.date(2022, 1, 1)
end_date = datetime.date(2025, 3, 23)

# List of target currencies
target_currencies = ['BND', 'MMK', 'KHR', 'IDR', 'LAK', 'MYR', 'PHP', 'SGD', 'THB', 'VND']

# List to store daily dataframes
df_list = []

# Fetch daily data
for single_date in tqdm(list(daterange(start_date, end_date)), desc="Fetching daily currency data"):
    date_str = single_date.strftime('%Y-%m-%d')
    try:
        url = f'https://www.xe.com/currencytables/?from=USD&date={date_str}'
        tables = pd.read_html(url)
        if tables and len(tables) > 0:
            daily_df = tables[0]
            daily_df['Date'] = date_str
            df_list.append(daily_df)
        
        # time.sleep(0.1)  # delay to avoid blocking
        
    except Exception as e:
        print(f"Error fetching data for {date_str}: {e}")
        time.sleep(10)

# Concatenate all daily data
if df_list:
    combined_df = pd.concat(df_list, ignore_index=True)

    # Verify currency column name
    currency_column = 'Currency Code' if 'Currency Code' in combined_df.columns else 'Currency'

    # Filter currencies
    filtered_df = combined_df[combined_df[currency_column].isin(target_currencies)]

    # Ensure 'Units per USD' column exists and convert to numeric
    value_column = 'Units per USD' if 'Units per USD' in filtered_df.columns else 'Rate'
    filtered_df[value_column] = pd.to_numeric(filtered_df[value_column], errors='coerce')

    # Convert 'Date' to datetime and calculate monthly average
    filtered_df['Date'] = pd.to_datetime(filtered_df['Date'])
    filtered_df['YearMonth'] = filtered_df['Date'].dt.to_period('M')

    # Compute monthly average
    monthly_avg = filtered_df.groupby(['YearMonth', currency_column])[value_column].mean().reset_index()
    monthly_avg.columns = ['YearMonth', 'Currency', 'AverageRate']

    # Save the monthly average data
    monthly_avg.to_csv('/Users/lishuyao/Documents/NUS/MODS/Y3S2/BT4301/tourism-consultancy/extracted_data/southeast_asia_currency_monthly_avg.csv', index=False)

    print(f"Monthly averages computed and saved to CSV.")
else:
    print("No data was collected")
```
