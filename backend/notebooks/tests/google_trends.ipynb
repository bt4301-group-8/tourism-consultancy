```python
import pandas as pd
import json
from datetime import datetime

# Replace 'results' with your actual JSON data (the provided data).
data = results

# Initialize an empty dictionary to temporarily store Singapore values per month-year
sg_values = {}
rows = []

# Function to convert timestamp to month-year
def timestamp_to_month_year(ts):
    return datetime.utcfromtimestamp(ts).strftime('%Y-%m')

# First pass: Collect all Singapore values per month-year to average later
for entry in data:
    tasks = entry.get('tasks', [])
    for task in tasks:
        for result in task.get('result', []):
            for item in result.get('items', []):
                keywords = item.get('keywords', [])
                singapore_index = keywords.index('Singapore')
                for data_point in item.get('data', []):
                    month_year = timestamp_to_month_year(data_point['timestamp'])
                    singapore_value = data_point['values'][singapore_index]
                    sg_values.setdefault(month_year, []).append(singapore_value)

# Compute average Singapore value per month-year
avg_sg_values = {month: sum(vals) / len(vals) for month, vals in sg_values.items()}

# Second pass: Normalize data points based on average Singapore value per month-year
for entry in data:
    tasks = entry.get('tasks', [])
    for task in tasks:
        for result in task.get('result', []):
            for item in result.get('items', []):
                keywords = item.get('keywords', [])
                for data_point in item.get('data', []):
                    month_year = timestamp_to_month_year(data_point['timestamp'])
                    singapore_avg_value = avg_sg_values[month_year]
                    for idx, country in enumerate(keywords):
                        country_value = data_point['values'][idx]
                        rows.append({
                            'month_year': month_year,
                            'country': country,
                            'value': (country_value / singapore_avg_value) * singapore_avg_value
                        })

# Create DataFrame and drop duplicate entries
df = pd.DataFrame(rows)
df.drop_duplicates(subset=['month_year', 'country'], keep='first', inplace=True)

# Sort DataFrame chronologically by month_year
df['month_year'] = pd.to_datetime(df['month_year'])
df = df.sort_values(by=['month_year', 'country']).reset_index(drop=True)
df['month_year'] = df['month_year'].dt.strftime('%Y-%m')
print(df)
```
